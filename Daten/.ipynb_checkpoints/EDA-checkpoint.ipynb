# Explorative Datenanalyse

```python
# ----------------------------
# Daten laden
# ----------------------------
import pandas as pd
df = pd.read_csv("merged_df.csv",  sep=';', low_memory=False)
#"C:\Users\sha_r\OneDrive - FH Graubünden\FHGR_\4. Semester\Masterthesis\dahsboard\data\merged_df.csv"

# ----------------------------
# Daten verarbeiten
# ----------------------------
df['time'] = pd.to_datetime(df['time'], errors='coerce')

df.rename(columns={
    'Temp_Mittel': 'Temperatur_Mittelwert',
    'Temp_Abw': 'Temperatur_Abweichung',
    'Niederschlag_Max_10min': 'Niederschlag_Max_10min',
    'Niederschlag_Tag': 'Niederschlag_Tagessumme',
    'Sonnenschein_h': 'Sonnenscheindauer_Stunden',
    'Wind_kmh': 'Windgeschwindigkeit_kmh',
    'Bodentemp_100cm': 'Bodentemperatur_100cm',
    'Schweregrad' : 'Schadenskosten'    
}, inplace=True)
```

```python
# Dimensionen des DataFrames
print(f"Form des DataFrames: {df.shape}")

# Datentypen und fehlende Werte
df.info()
```

```python
# Statistiken für numerische Werte
df.describe().T
```

```python
# Statistiken für kategorische Spalten
df.select_dtypes(include='object').describe().T
```

```python
# Gültige Zeitwerte prüfen
df['time'].min(), df['time'].max()

# Zeitliche Verteilung der Datenpunkte
df['time'].dt.year.value_counts().sort_index().plot(kind='bar', title='Anzahl Datenpunkte pro Jahr')
```

```python
# Wie viele Stationen?
print("Anzahl Wetterstationen:", df['Station'].nunique())
print("Stationen:", df['Station'].unique())

# Datenpunkte pro Station
df['Station'].value_counts().plot(kind='bar', title='Datenpunkte pro Station')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Nur numerische Spalten für Korrelation
num_cols = df.select_dtypes(include='number')

plt.figure(figsize=(10, 6))
sns.heatmap(num_cols.corr(), annot=True, cmap='coolwarm')
plt.title('Korrelationsmatrix')

# Grafik speichern
plt.savefig('korrelationsmatrix.png', dpi=300, bbox_inches='tight')
print("Korrelationsmatrix wurde als 'korrelationsmatrix.png' gespeichert.")

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import os

# Variablenliste
cols = [
    'Temperatur_Mittelwert',
    'Temperatur_Abweichung',
    'Niederschlag_Max_10min',
    'Niederschlag_Tagessumme',
    'Sonnenscheindauer_Stunden',
    'Windgeschwindigkeit_kmh',
    'Bodentemperatur_100cm',
    'Schadenskosten',
    'Unwetterart'
]

# Kategorische Variablen
kategorisch = ['Schadenskosten', 'Unwetterart']

# Plot-Matrix
n_cols = 3
n_rows = (len(cols) + n_cols - 1) // n_cols

plt.figure(figsize=(n_cols * 5, n_rows * 4))
sns.set(style="whitegrid")

for i, col in enumerate(cols, 1):
    plt.subplot(n_rows, n_cols, i)
    
    if col in kategorisch:
        df[col].dropna().value_counts().plot(kind='bar', color='coral')
        plt.title(f'Häufigkeit: {col}')
        plt.xlabel(col)
        plt.ylabel('Anzahl')
        plt.xticks(rotation=45)
    else:
        sns.histplot(df[col].dropna(), kde=True, bins=30, color='steelblue')
        plt.title(f'Verteilung: {col}')
        plt.xlabel(col)
        plt.ylabel('Anzahl')

plt.tight_layout()

# 🔽 Grafik im Skript-Ordner speichern
save_path = os.path.join(os.getcwd(), 'verteilungen.png')
plt.savefig(save_path, dpi=300)
print(f"Grafik gespeichert unter: {save_path}")

plt.show()
```
